# Analysing the clean data [ Data analaysis and visualtions]

## Reservation status count

```python
df['reservation_status'].value_counts().plot(kind='bar')
plt.title('Reservation Status Count')
plt.show()
```

## Reservation status in hotels

```python
plt.figure(figsize=(8,6))
sns.countplot(x='hotel', hue='is_canceled', data=df)
plt.title('Reservation status in hotels')
plt.show()
```

```python
resort_hotel = df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

## Reservation status per month

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (8,6))
sns.countplot(x = 'month', hue= 'is_canceled', data = df)
plt.title('Reservationa status per month')
plt.xlabel('month')
plt.ylabel(' number of reservation')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

## Average daily reservation per month

```python
plt.figure(figsize=(13,8))
plt.title('ADR per month', fontsize=20)
sns.barplot(x='arrival_date_month', y='adr', data=df.groupby('arrival_date_month')['adr'].sum().reset_index())
plt.show()
```

## Top 10 countries by reservation

```python
cancelled_data = df[df['is_canceled']== 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 country with reservation canceled')
plt.pie(top_10_country.values, labels=top_10_country.index, autopct='%1.1f%%')
plt.show()
```
